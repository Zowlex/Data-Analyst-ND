```python
import requests
from PIL import Image
from io import BytesIO

def get_file(url, save=True):
    """
    Downloads and save the file from the provided url and save to data/file_name,
    if save is False it returns the img
    """
    response = requests.get(url)

    file_name = url.split('/')[-1]
    
    if save:
        try:
            with open(f'{path}/{file_name}', mode='wb') as file:
                file.write(response.content)
                print(f'{path}/{file_name} was written successfully...')
        except requests.exceptions.HTTPError as err:
            raise SystemExit(err)
    else:
        img = Image.open(BytesIO(response.content))
        return img
```

- Scrape likes, retweet counts and additional data for each tweet_id present in the tweet archive above using Tweepy

```python
#read keys from env vars
consumer_key = os.getenv('CONSUMER_KEY')
consumer_secret = os.getenv('CONSUMER_SECRET')
access_token = os.getenv('ACCESS_TOKEN')
access_secret = os.getenv('ACCESS_SECRET')

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
file_name = 'tweet_json.txt'
deleted_ids = []
#write each tweet to tweet_json.txt

if os.path.isfile(f'{path}/{file_name}'):
    pass
else:
    #download
    with open(f'{path}/{file_name}', 'w') as file: 
        for tweet_id in tqdm(twitter_archive_df.tweet_id.values):
            try:
                tweet = api.get_status(tweet_id,tweet_mode='extended')
            except:
                deleted_ids.append(tweet_id)

            tweet_str = json.dumps(tweet._json)
            file.write(tweet_str+"\n")
```

```python
print(f'number of deleted tweets: {len(deleted_ids)}')
```

```python
#create dataframe from tweet_json file
tweet_json_df = pd.DataFrame()


with open(f'{path}/tweet_json.txt') as file:
    for line in tqdm(file):
        tweet = json.loads(line)
        tweet_json_df = tweet_json_df.append({'tweet_id': tweet['id_str'],
                                              'retweet_count': tweet['retweet_count'],
                                              'favorite_count': tweet["favorite_count"]
                                            }, ignore_index=True)
```

#### Define

- fix numerators and denominators by extracting values with decimals from text and removing outliers

#### Code

```python
def remove_outliers(col):
    """
    remove outliers by selecting values between q_low and q_high
    """
    q_low = col.quantile(0.01)
    q_high = col.quantile(0.99)
    
    return col[(col>=q_low)&(col<=q_high)]
```

```python
#extract numerator and denominator from text column
extracted_values = twitter_archive_df.text.str.extract('((?:\d+\.)?\d+)\/((\d+\.)?\d+)', expand=True)

#get numerator
numerator = extracted_values.loc[:,0].astype(float)
#get denominator
denominator = extracted_values.loc[:,1].astype(float)

#set values to df
twitter_arch_clean.rating_numerator = remove_outliers(numerator)
twitter_arch_clean.rating_denominator = remove_outliers(denominator)
```

#### Define

- group dog stages into one column by combining different columns

#### Code

```python
def merge_dogstages(df):
    """
    replace every missing and None dog_stage with empty space and combine 
    different dog stages together for each row.
    """
    cols = ['doggo', 'floof', 'pupper', 'puppo']
    for col in cols:
        df[col] = df[col].replace('None', '')
        df[col] = df[col].replace(np.NaN, '')
    df['stage'] = df.doggo + df.floof + df.pupper + df.puppo
    df.loc[df.stage == 'doggopupper', 'stage'] = 'doggo, pupper'
    df.loc[df.stage == 'doggopuppo', 'stage'] = 'doggo, puppo'
    df.loc[df.stage == 'doggofloof', 'stage'] = 'doggo, floof'
    df['stage'] = df['stage'].replace('', 'None')
    return df.drop(cols, axis=1)
```

```python
twitter_arch_clean = merge_dogstages(twitter_arch_clean)
```

## Predict missing dog_stages

We will try to use the data we have and try to predict the missing dog_stages, so let's give it a try

```python
df = pd.read_csv('data/clean/twitter_archive_master.csv')
```

```python
#save indx to use later
indx = df.index
```

there are 1665 values to predict of four classes, so our problem is a multiclass classification problem

```python
#let's define our train and test data
train = df[df.stage != 'None']
test = df[df.stage == 'None']
print(f'train: {train.shape}, test: {test.shape}')
```

```python
#oversampling 
from imblearn.over_sampling import SMOTE

#drop rows with class of value count = 1 
single_class = train[(train.stage=='doggo, floof')|(train.stage=='doggo, puppo')]
to_drop = single_class.index
train_le = train.drop(to_drop)


stages = pd.factorize(train.stage)[1]

for col in train_le.select_dtypes(include='object').columns:
    train_le[col] = pd.factorize(train_le[col])[0]


    
smote = SMOTE(sampling_strategy='auto', n_jobs=-1)
X_sm, y_sm = smote.fit_resample(train_le.drop('stage', axis=1), train_le['stage'])

df_oversampled = pd.DataFrame(X_sm, columns=train_le.drop('stage', axis=1).columns)
df_oversampled['target'] = y_sm

#df_oversampled['target'].value_counts().plot(kind='bar', title='Count (target)');
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split , StratifiedKFold

features_to_drop = ['target','text','source', 'img_url', 'rating_denominator', 'expanded_urls']

features = set(df_oversampled.columns) - set(features_to_drop)

X = df_oversampled[features]
y = df_oversampled.target

#label encode test
test_le = test.copy()
for col in test_le.select_dtypes(include='object').columns:
    test_le[col] = pd.factorize(test_le[col])[0]
    
    
skf = StratifiedKFold(n_splits=5)

preds = np.zeros((len(test),5), dtype=object)
scores = []

seed = 42

lgb_clf = lgb.LGBMClassifier(random_state=seed)

for i, (train_index, val_index) in enumerate(tqdm(skf.split(X,y))):
    x_train, x_val = X.loc[train_index], X.loc[val_index]
    y_train, y_val = y.loc[train_index], y.loc[val_index]
    lgb_clf.fit(x_train, y_train)
    pred = lgb_clf.predict(x_val)
    score = accuracy_score(y_val, pred)
    print(f'fold{i+1} accuracy: {score}')
    scores.append(score) 
    preds[:,i] = lgb_clf.predict(test_le[features]).squeeze()
print(f'Avg Accuracy: {np.mean(scores)}')
```

I used a 5 fold cross validation to predict the stage and saved the results to preds to use the pick the mode of each prediction, meaning for each single value we predict 5 times and get the most occured pred.

```python
from scipy.stats import mode

predictions = mode(preds, axis=1).mode
```

```python
#assign to stage
test.stage = predictions

#restore original values
def original_values(x):
    if x==0:
        x=stages[0]
    elif x==1:
        x=stages[1]
    elif x==2:
        x=stages[2]
    elif x==3:
        x=stages[3]
    else:
        x=stages[4]
    return x
test.stage = test.stage.apply(original_values)
```

```python
#concat results
df = pd.concat([train, test]).set_index(indx)
```
